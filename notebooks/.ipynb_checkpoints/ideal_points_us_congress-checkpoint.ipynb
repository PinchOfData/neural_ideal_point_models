```python
import pandas as pd
df = pd.read_csv(
    '/home/germain/Documents/topic models/generalized_topic_model_dev/data/us_congressional_record/us_congress_speeches_daily_processed.csv'
)
df = df.sample(50000)
df.dropna(subset=['doc_clean'], inplace=True)
df['word_count'] = df['doc_clean'].apply(lambda x: len(x.split()))
df = df[df["word_count"] > 30]
df = df.reset_index()
df
```

```python
import sys
sys.path.append('../IdealPointNN/')
from corpus import IdealPointCorpus
from ideal_point_model import IdealPointNN

stopwords = ["yield", "second", "minute", "amendment", "yea", "nay", "gentleman", "gentlewoman", "thing", "conference", "conferee",
             "thank", "vote", "quorum", "majority", "vote", "bill", "object", "unanimous", "consent", "distinguished", 
             "debate", "call", "rollcall", "ask", "provision", "title", "time", "rise", "support", "morning", "noon", "today", "tomorrow",
             "chairman", "resolution", "say", "year", "make", "think", "colleague", "today", "b2", "report",
             "committee", "want", "need", "go", "get", "other", "member", "program", "know", "take", 
             "waive", "clause", "section", "rule", "tell", "talk", "consideration", "legislation", "provide",
             "issue", "re", "ve", "let", "subsection", "adjourn", "adjournment", "motion", "objection", 
             "question", "session", "order", "proposition", "oclock", "matter", "point", "germane", "germaneness",
             "unanimousconsent", "ratification", "senator", "cloture", "subjectmatter", "designee", "agree", "disagree"]

train_dataset = IdealPointCorpus(
    df,  
    prevalence = None,
    content = "~ 1",
    #labels = "~party",
    vectorizer_args={'ngram_range':(1, 2), 'max_df':0.99, 'min_df':0.01, 'stop_words': stopwords}
)
```

```python
#train_dataset.M_labels = train_dataset.M_labels[:,-1]
#train_dataset.M_labels = train_dataset.M_labels.reshape(train_dataset.M_labels.shape[0], -1)
```

```python
# Train the model
tm = IdealPointNN(
    train_dataset, 
    n_dims=2,
    update_prior=False,
    encoder_hidden_layers=[],
    decoder_hidden_layers=[],
    log_every_n_epochs = 1,
    print_every_n_batches = 100,
    w_pred_loss = 100000,
    predictor_type = 'classifier'
)
```

```python
import torch
import torch.nn.functional as F
l_content_covariates = []
topK = 8
tm.AutoEncoder.eval()
with torch.no_grad():
    dim_words = {}
    #values_column1 = torch.arange(-1, 1, step=0.1,dtype=torch.float)
    #values_column2 = torch.zeros_like(values_column1)
    #new_rows = torch.stack((values_column1, values_column2), dim=1)
    #idxes = new_rows.to(tm.device)
    idxes = torch.tensor([[-1,0,0],[1,0,0],[0,-1,0],[0,1,0]], dtype = torch.float).to(tm.device)
    word_dist = tm.AutoEncoder.decode(idxes)
    _, indices = torch.topk(word_dist, topK, dim=1)
    indices = indices.cpu().tolist()
    for dim_value in range(4):
        dim_words["Value_{}".format(dim_value)] = [
            tm.id2token[idx] for idx in indices[dim_value]
        ]

print(
    "\n".join(
        [
            "{}: {}".format(k, str(v))
            for k, v in dim_words.items()
        ]
    )
)
```
